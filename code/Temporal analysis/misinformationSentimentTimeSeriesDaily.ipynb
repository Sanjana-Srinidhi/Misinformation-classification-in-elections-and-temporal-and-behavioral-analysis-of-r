```python
import re
import pandas as pd
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
import matplotlib.patches as mpatches
```

```python
misinformation_df = pd.read_csv('/content/drive/MyDrive/DA/data/Time Series/misinformation_timeseries_daily.csv')
misinformation_df.head()
```

```python
column = 'Frequency'
misinformation_df[column] = (misinformation_df[column] - misinformation_df[column].min()) / (misinformation_df[column].max() - misinformation_df[column].min())  
misinformation_df.head()
```

```python
sentiment_df = pd.read_csv('/content/drive/MyDrive/DA/data/Time Series/negative_sentiment_time_series_daily.csv')
sentiment_df.head()
```

```python
column = 'Frequency'
sentiment_df[column] = (sentiment_df[column] - sentiment_df[column].min()) / (sentiment_df[column].max() - sentiment_df[column].min())  
misinformation_df.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = misinformation_df['Time'], y = misinformation_df['Frequency'], name = 'Misinformation rate', line = dict(color = 'firebrick', width = 2)))
fig.add_trace(go.Scatter(x = sentiment_df['Time'], y = sentiment_df['Frequency'], name = 'Negative sentiment rate', line = dict(color = 'royalblue', width = 2)))
fig.update_layout(xaxis_title = 'Month-Year', yaxis_title = '')
```
