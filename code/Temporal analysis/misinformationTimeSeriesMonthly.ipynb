```python
import re
import pandas as pd
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
import matplotlib.patches as mpatches
```

```python
df = pd.read_csv('/content/drive/MyDrive/DA/data/combined_dataset.csv')
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], errors = 'coerce')
df.dtypes
```

```python
misinformation_df = df.loc[df['label'] == 0]
misinformation_df = misinformation_df[['date', 'text', 'label']]
misinformation_df.head()
```

```python
def timec(i):
  i = str(i)
  i = i[0:7]
  return i
misinformation_df['date'] = misinformation_df['date'].apply(timec)
misinformation_df.head()
```

```python
list_t = misinformation_df['date'].values.tolist()
misinformation_dict = dict()
for i in range(len(list_t)):
  if list_t[i] not in misinformation_dict:
    misinformation_dict[list_t[i]] = 0
  misinformation_dict[list_t[i]] = misinformation_dict[list_t[i]] + 1
print(misinformation_dict)
list_t = list(misinformation_dict.keys())
count = list(misinformation_dict.values())
time = pd.DataFrame({'Time': list_t, 'Frequency': count})
time = time.dropna()
time.reset_index(inplace = True)
time = time.sort_values('Time')
time.head()
```

```python
time["Time"]  = pd.to_datetime(time["Time"] , format='%Y-%m', errors = 'coerce')
time["Time"] = time["Time"].dt.strftime('%B-%Y')
time.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = time['Time'], y = time['Frequency'], name = 'Misinformation news', line = dict(color = 'blue', width = 2)))
fig.update_layout(xaxis_title = 'Month-Year', yaxis_title = 'Frequency')
fig.show()
```

```python
time.to_csv('/content/drive/MyDrive/DA/data/misinformation_timeseries_monthly.csv', index = False)
```
