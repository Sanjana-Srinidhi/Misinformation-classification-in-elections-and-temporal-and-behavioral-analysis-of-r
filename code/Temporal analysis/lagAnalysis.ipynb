```python
import re
import pandas as pd
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
import matplotlib.patches as mpatches
from scipy import signal
```

```python
misinformation_df = pd.read_csv('misinformation_timeseries_daily.csv')
misinformation_df.head()
```

```python
column = 'Frequency'
misinformation_df[column] = (misinformation_df[column] - misinformation_df[column].min()) / (misinformation_df[column].max() - misinformation_df[column].min())  
misinformation_df.head()
```

```python
# sentiment_df = pd.read_csv('/content/drive/MyDrive/DA/data/Time Series/negative_sentiment_time_series_daily.csv')
sentiment_df = pd.read_csv('negative_sentiment_time_series_daily.csv')
sentiment_df.head()
```

```python
column = 'Frequency'
sentiment_df[column] = (sentiment_df[column] - sentiment_df[column].min()) / (sentiment_df[column].max() - sentiment_df[column].min())  
misinformation_df.head()
```

```python
misinformation = misinformation_df['Frequency'].tolist()
sentiment = sentiment_df['Frequency'].tolist()
```

```python
a = misinformation - np.mean(misinformation)
b = sentiment - np.mean(sentiment)
correlation = signal.correlate(a, b, mode = 'full')
lags = signal.correlation_lags(len(misinformation), len(sentiment), mode = 'full')
lag = lags[np.argmax(abs(correlation))]
print(lag)
```
