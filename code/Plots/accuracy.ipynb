```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'XGBoost': 0.98, 'Logistic Regression': 0.72, 'KNN': 0.64, 'Decision Trees': 0.55, 'Random Forests': 0.54, 'Linear SVM': 0.53, 'Adaboost': 0.50, 'Passive Aggressive Classifier': 0.50, 'Naive Bayes Classifier': 0.50}
models = list(data.keys())
acc = list(data.values())
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(models, acc, color = 'blue', width = 0.4)
plt.xticks(rotation = 90)
plt.xlabel("ML Models")
plt.ylabel("Accuracy")
plt.show()
```
